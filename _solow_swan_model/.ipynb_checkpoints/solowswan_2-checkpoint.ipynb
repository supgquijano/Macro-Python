```python
import numpy as np
import matplotlib.pyplot as plt

# numpy es para realizar operaciones matemáticas.
# matplotlib.pyplot es para realizar gráficos.

# RENDIMIENTOS MARGINALES POSITIVOS Y DECRECIENTES
# CONDICIONES DE INADA

# Parámetros de la función Cobb-Douglas
alpha = 0.85  # Parámetro de capital
beta = 1 - alpha   # Parámetro de trabajo

# Definición de la función Cobb-Douglas
def cobb_douglas(K, L):
    return K**alpha * L**beta

# Definición de los rendimientos marginales
def rendimiento_marginal_capital(K, L):
    return alpha * K**(alpha - 1) * L**beta

def rendimiento_marginal_trabajo(K, L):
    return beta * K**alpha * L**(beta - 1)

# Generación de datos
K_values = np.linspace(0.1, 100, 500)
L_values = np.linspace(0.1, 100, 500)

# Creación de la figura con subgráficos
fig, axs = plt.subplots(1, 4, figsize=(12, 5))

# Gráfico de la función Cobb-Douglas con Ls constantes.
axs[0].plot(K_values, cobb_douglas(K_values, 1), label='L = 1')
axs[0].plot(K_values, cobb_douglas(K_values, 2), label='L = 2')
axs[0].plot(K_values, cobb_douglas(K_values, 3), label='L = 3')
axs[0].set_xlabel('Capital (K)')
axs[0].set_ylabel('Producción')
axs[0].set_title('Función Cobb-Douglas')
axs[0].legend()

# Gráfico de la función Cobb-Douglas con Ks constantes.
axs[1].plot(L_values, cobb_douglas(1, L_values), label='K = 1')
axs[1].plot(L_values, cobb_douglas(2, L_values), label='K = 2')
axs[1].plot(L_values, cobb_douglas(3, L_values), label='K = 3')
axs[1].set_xlabel('Trabajo (L)')
axs[1].set_ylabel('Producción')
axs[1].set_title('Función Cobb-Douglas')
axs[1].legend()

# Gráfico de los rendimientos marginales de K con L constante.
axs[2].plot(K_values, rendimiento_marginal_capital(K_values, 1), label='Rendimiento Capital, L = 1')
axs[2].set_xlabel('Capital (K)')
axs[2].set_ylabel('Rendimientos Marginales')
axs[2].set_title('Rendimientos Marginales')
axs[2].legend()

# Gráfico de los rendimientos marginales de L con K constante.
axs[3].plot(L_values, rendimiento_marginal_trabajo(1, L_values), label='Rendimiento Trabajo, K = 1')
axs[3].set_xlabel('Trabajo (L)')
axs[3].set_ylabel('Rendimientos Marginales')
axs[3].set_title('Rendimientos Marginales')
axs[3].legend()

plt.tight_layout()
plt.show()
```
